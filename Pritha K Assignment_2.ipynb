**Preprocessing categorical data using the Onehot encoder**

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
```

**Separating categorical columns from the dataset**

```python
categorical_columns = [cname for cname in df.columns if df[cname].dtype == "object"]
numerical_columns   = [cname for cname in df.columns if df[cname].dtype in ['int64','float64']]
```

```python
categorical_columns = df[categorical_columns]
numerical_columns   = df[numerical_columns]
```

```python
categorical_columns.head()
```

```python
numerical_columns.head()
```

**Drop the surname, Rownumber, customerId from the categorical and numerical columns**

```python
categorical_columns = categorical_columns.drop('Surname',axis=1)
numerical_columns   = numerical_columns.drop(['RowNumber','CustomerId'],axis=1)
```

```python
categorical_columns.head()
```

```python
encoded_categorical = pd.DataFrame(ordinal_encoder.fit_transform(categorical_columns))
```

```python
encoded_categorical
```

```python
encoded_categorical.columns = categorical_columns.columns
```

```python
encoded_categorical
```

**Concatenate preprocessed categorical and the numerical data**

```python
df1 = pd.concat([numerical_columns,encoded_categorical],axis=1)
```

```python
df1.head()
```

**Split it dependent and independent**

```python
X = df1.drop('Exited',axis=1)
y = df1['Exited']
```

**Scale the independent data using StandardScalar**

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
scaled_x = pd.DataFrame(scalar.fit_transform(X))
```

```python
scaled_x
```

**Split the dependent and independent data**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(X_train,y_train)
```

```python
y_pred = model1.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print("classification_report is: ",classification_report(y_pred,y_test))
print("confusion_matrix is: ",confusion_matrix(y_pred,y_test))
print("accuracy_score is: ", accuracy_score(y_pred,y_test))
```

```python
model1.score(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```
